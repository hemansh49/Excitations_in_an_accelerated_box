```python
import numpy as np
from mpmath import gammainc

def w(m):
    return m*np.pi
def o(n,a):
    return n*np.pi*a/np.log(1+a)

def B(n,m,a):
    ww = w(m)
    oo = o(n,a)
    t1 = 1/np.sqrt(a*np.log(1+a))
    t2 = np.sqrt(oo/ww)
    t3 = np.exp(-np.pi*oo/(2*a))
    t4 = np.exp(1j*(oo/a)*np.log(ww/a))
    t5 = np.exp(1j*ww/a)
    t6 = gammainc(-1j*oo/a, 1j*ww/a, 1j*ww/a + 1j*ww)
    t7 = np.real(t4*t5*t6)
    return t1*t2*t3*t7

def P(n,a,M):
    p = 0
    for m in range(1,M+1):
        p = p + np.power(B(n,m,a),2)
    return p
```

```python
a_arr = np.array([0.25, 1.0, 10.0])
n_range = np.arange(1, 101)

# Preallocate a list to store each row (corresponding to each `a`)
P_spectrum_list = []

for a in a_arr:
    P_row = [P(n, a, 20000) for n in n_range]  # shape (100,)
    P_spectrum_list.append(P_row)

# Convert list of lists into a NumPy array of shape (len(a_arr), len(n_range))
P_spectrum = np.array(P_spectrum_list)
```

```python
np.save("P_spectrum_3.npy", P_spectrum)
np.save("a_arr_3.npy", a_arr)
np.save("n_range_3.npy", n_range)
```
