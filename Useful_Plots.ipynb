```python
# Setup
a = 0.4
t_arr = np.linspace(-10, 5.5, 100)

Lx, Ly, Rx, Ry = [], [], [], []

for t in t_arr:
    if t <= 0:
        Lx.append(0)
        Ly.append(t)
        Rx.append(2)
        Ry.append(t)
    else:
        Lx.append((1/a) * np.cosh(a*t) - (1/a))
        Ly.append((1/a) * np.sinh(a*t))
        Rx.append((2 + 1/a) * np.cosh(a*t) - (1/a))
        Ry.append((2 + 1/a) * np.sinh(a*t))

# Plot
fig, ax = plt.subplots(figsize=(4, 8), dpi=300)

ax.plot(Lx, Ly, '-.', label='Left edge')
ax.plot(Rx, Ry, '--', label='Right edge')

# Set fixed range
ax.set_xlim(-2, 10)
ax.set_ylim(-10, 10)

# Move axes to center
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')

# Hide all spines
for spine in ['left', 'bottom', 'top', 'right']:
    ax.spines[spine].set_visible(False)

# Draw X-axis arrow
from matplotlib.patches import FancyArrowPatch
arrow_style = dict(arrowstyle='->', mutation_scale=15, lw=1.8, color='black')

ax.add_patch(FancyArrowPatch(( -0.5, 0), (10, 0), **arrow_style))  # X-axis
ax.add_patch(FancyArrowPatch((0, -10), (0, 10), **arrow_style))  # T-axis

# Label axes at the end
ax.text(9.6, -0.7, f'$X$', fontsize=14)
ax.text(-0.6, 9.6, f'$T$', fontsize=14)

ax.set_xticks([])  # Removes x-axis numbers
ax.set_yticks([])  # Removes y-axis numbers

# General formatting
ax.set_aspect('equal')
ax.legend()
ax.grid(True, linestyle='--')
plt.tight_layout()
plt.show()
```

```python
def Peve(n, a, M):
    P = 0
    m = 1
    while m<=M:
        P = P + n*m/((n+m)**6)
        m = m + 2
    return (4*(a**2)/(np.pi**4))*P

def Podd(n, a, M):
    P = 0
    m = 2
    while m<=M:
        P = P + n*m/((n+m)**6)
        m = m + 2
    return (4*(a**2)/(np.pi**4))*P

def P_lowa(n, m, M):
    if n%2 == 0:
        return Peve(n, a, M)
    else:
        return Podd(n, a, M)

a = 1
N = 10
n_cont = np.linspace(0.5, N+0.5, 50)
Peve_arr = np.array([Peve(n, a, 400) for n in n_cont])
Podd_arr = np.array([Podd(n, a, 400) for n in n_cont])
n_range = np.arange(1, N+1)
P_lowa_arr = np.array([P_lowa(n, a, 400) for n in n_range])

plt.figure(figsize=(8, 6), dpi=300)

plt.plot(n_cont, Peve_arr, '--', c="green", label="even mode")
plt.plot(n_cont, Podd_arr, '-.', c="red", label="odd mode")
plt.plot(n_range, P_lowa_arr, 'o-', color = 'tab:blue', alpha = 0.5, label = "Actual")
plt.scatter(n_range, P_lowa_arr, color='tab:blue')
plt.xlabel("Mode", fontsize = 12)
plt.xlim(0, N+0.5)
plt.ylim(0, 0.00015)
plt.xticks(np.arange(1, N+1, 1))
plt.ylabel("$P(n)/a^2L^2 $", fontsize = 12)
plt.tick_params(axis='both', which='major', labelsize=12)
#plt.title("Low acceleration spectrum")
plt.grid(axis = 'x')
for spine in ['top', 'right']:
    plt.gca().spines[spine].set_visible(False)
plt.legend()
plt.show()
```

```python
A = 3.243

def p(n, m, y):
    return np.log(1+A*y)/np.log(1+A) + m/n * y

n, m = 100, 50
yarr = np.linspace(0, 1, 100)
pvals = p(n, m, yarr)

plt.figure(figsize=(8, 6), dpi=300)
plt.plot(yarr, pvals)

# Format axes
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Compute special points
p0 = p(n, m, 0)  # should be 0
p1 = p(n, m, 1)  # value at y=1

# Add grid lines at y=0,1 and p=0, p(1)
ax.axvline(0, color='gray', linestyle='--', linewidth=0.8)
ax.axvline(1, color='gray', linestyle='--', linewidth=0.8)
ax.axhline(p0, color='gray', linestyle='--', linewidth=0.8)
ax.axhline(p1, color='gray', linestyle='--', linewidth=0.8)

# Set ticks
ax.set_xticks([0, 1])
ax.set_yticks([0, p1])
ax.set_yticklabels([r"$0$", fr"$1+\frac{{m}}{{n}}$"])  # label p(1) as 1+m/n

ax.set_aspect('equal')

# Labels
plt.xlabel(r"$y$", fontsize = 12)
plt.ylabel(r"$p(y)$", fontsize = 12)
plt.tick_params(axis='both', which='major', labelsize=12)
plt.show()
```

```python
def ratio(n, T, aaa, L):
    f1 = np.exp(n*np.pi*3.1615e-26/(L*1.38065e-23*T)) - 1
    f2 = aaa**2 * L**2 /(((2.9979e8)**4 * n**3))
    return (1/(10*np.pi**4)) * f1 * f2

# Temperature range
T_arr = np.linspace(1, 125, 125)

ratio_100 = []
ratio_1000 = []
ratio_10000 = []

aaa = 100
L = 0.01

for T in T_arr:
    if T < 30:
        ratio_100.append(ratio(100, T, aaa, L))
        ratio_1000.append(ratio(1000, T, aaa, L))
    else:
        ratio_100.append(ratio(100, T, aaa, L))
        ratio_1000.append(ratio(1000, T, aaa, L))
        ratio_10000.append(ratio(10000, T, aaa, L))

ratio_100 = np.array(ratio_100)
ratio_1000 = np.array(ratio_1000)
ratio_10000 = np.array(ratio_10000)

# Create figure and axis
fig, ax = plt.subplots(figsize=(8, 6), dpi=300)

# Plot
# ax.plot(T_arr, ratio_100, label=r"$n = 100$")
ax.plot(T_arr, ratio_1000, '-.', label=r"$n = 1000$")
ax.plot(T_arr[29:], ratio_10000, '--', label=r"$n = 10000$")  # starts at T=30

# Add horizontal line at P_n : N_n = 10^5
ax.axhline(1e5, color="green", linestyle="-", linewidth=1.2, alpha = 0.5,
           label=r"$P_n : \mathcal{N}_n = 10^5$")

# Log scale
ax.set_yscale("log")

# Gridlines every 5 units on x-axis
ax.set_xticks(np.arange(0, 105, 5))
ax.grid(True, which="both", axis="x", linestyle="--", alpha=0.7)
ax.grid(True, which="both", axis="y", linestyle="--", alpha=0.7)

# Hide top and right spines
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Limits
ax.set_xlim(0, 80)
ax.set_ylim(1e-45, 1e45)

# Labels & legend
ax.set_xlabel("Temperature (K)", fontsize=14)
ax.set_ylabel(r"$P_n \,:\, \mathcal{N}_n$", fontsize=14)
ax.tick_params(axis='both', which='major', labelsize=12)
ax.legend(fontsize=12)

plt.show()
```
